```python
# CELL: "What the hell are the real columns in GDELT v2 events table in 2025?"
from google.cloud import bigquery
import os
from dotenv import load_dotenv
load_dotenv()

client = bigquery.Client(project=os.getenv("GOOGLE_CLOUD_PROJECT"))

# This query returns the actual schema
schema_query = """
SELECT column_name, data_type
FROM `gdelt-bq.gdeltv2`.INFORMATION_SCHEMA.COLUMNS
WHERE table_name = 'events'
ORDER BY ordinal_position
"""

schema = client.query(schema_query).to_dataframe()
print("Real GDELT v2.events columns (as of Nov 2025):")
print(schema[['column_name', 'data_type']].to_string(index=False))
```
